```python
import numpy as np

arr = np.array([[6, -8, 73, -110],
                [np.nan, -8, 0, 94]])

# Replace NaN with 0
arr = np.nan_to_num(arr, nan=0)

print("After replacing NaN with 0:")
print(arr)

# Interchange rows and columns (Transpose)
transposed = arr.T

print("\nAfter Interchanging Rows and Columns:")
print(transposed)
```

```python
import numpy as np

arr = np.array([[10, np.nan, 30],
                [20, 40, np.nan],
                [30, 50, 60]])

col_mean = np.nanmean(arr, axis=0)

inds = np.where(np.isnan(arr))
arr[inds] = np.take(col_mean, inds[1])

print("Array after replacing NaN with column average:")
print(arr)
```

```python
import numpy as np

arr1 = np.array([[3, 4],
                 [5, 6]])

arr2 = np.array([[1, 0],
                 [7, 8]])

# Average
avg = (arr1 + arr2) / 2

print("Average Array:")
print(avg)

# Combine arrays
combined = np.concatenate((arr1.flatten(),
                           arr2.flatten()))

# Mean
mean_val = np.mean(combined)

# Median
median_val = np.median(combined)

# Mode
values, counts = np.unique(combined, return_counts=True)
mode_val = values[np.argmax(counts)]

print("\nMean =", mean_val)
print("Median =", median_val)
print("Mode =", mode_val)
```

```python
import numpy as np

A = np.array([[1, -2, 3],
              [-1, 3, -1],
              [2, -5, 5]])

B = np.array([9, -6, 17])

# Method 1: linalg.solve()
solution = np.linalg.solve(A, B)

print("Using linalg.solve():")
print(solution)

# Method 2: Inverse Matrix
A_inv = np.linalg.inv(A)

result = np.dot(A_inv, B)

print("\nUsing Inverse Matrix:")
print(result)
```

```python
import matplotlib.pyplot as plt
import numpy as np

subjects = ['Python', 'DBMS', 'AI', 'CN', 'OS']

sem1 = [75, 70, 80, 65, 72]
sem2 = [85, 80, 88, 78, 82]

x = np.arange(len(subjects))
width = 0.35

plt.figure(figsize=(8,5))

plt.bar(x - width/2, sem1,
        width=width,
        label='Semester 1')

plt.bar(x + width/2, sem2,
        width=width,
        label='Semester 2')

plt.title("Semester Result Comparison")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.xticks(x, subjects)
plt.legend()
plt.grid(True)

plt.show()
```
