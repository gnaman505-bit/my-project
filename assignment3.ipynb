```python
def maximum(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

result = maximum(num1, num2, num3)

print("Maximum number is:", result)
```

```python
def distinct_list(lst):
    new_list = []

    for item in lst:
        if item not in new_list:
            new_list.append(item)

    return new_list

numbers = [1, 2, 2, 3, 4, 4, 5, 1]

result = distinct_list(numbers)

print("Original List:", numbers)
print("Distinct List:", result)
```

```python
def multiply(list):
    result=1

    for num in list:
        result = result * num    
        
    return result
    
numbers = [4,5,3,2,7,8]    

product = multiply(numbers)

print("multiplication of the elements are",product)
```

```python
def factorial(n):
    result=1

    for i in range(1,n+1):
        result = result*i
    return result

num = int(input("enter a number:"))

fact = factorial(num)

print("the factorial of",num," is ",fact)
    
```

```python
def is_prime(n):

    if n <= 1:
        print("Not a prime number")
        return

    for i in range(2, n):
        if n % i == 0:
            print("Not a prime number")
            return
        
    print("Prime number")

num = int(input("Enter a number: "))

is_prime(num)
```

```python
def count_case(text):

    upper = 0
    lower = 0

    for ch in text:

        if ch.isupper():
            upper += 1

        elif ch.islower():
            lower += 1

    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)

string = input("Enter a string: ")

count_case(string)
```
