```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error! Division by zero is not allowed."
    return a / b

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Addition =", add(num1, num2))
print("Subtraction =", subtract(num1, num2))
print("Multiplication =", multiply(num1, num2))
print("Division =", divide(num1, num2))
```

```python
num = input("enter any number: ")
reverse = num[-1::-1]

if num==reverse:
    print(num,"number is palindrome")
else:
    print("not an palindrome")    
```

```python
marks = {
    "Math": 95,
    "Physics": 90,
    "Chemistry": 92
}

print(marks.keys())

print(marks.values())

print(marks.get("Math"))

print(marks.items())

print("Math Marks:", marks["Math"])

for subject, score in marks.items():
    print(subject, "=", score)
```

```python
#tuples
marks = (85, 90, 78, 92)

print("First mark:", marks[0])
print("Total marks:", sum(marks))
print("Highest mark:", max(marks))
```

```python
fruits = {"apple", "banana"}
fruits.add("mango")
print(fruits)

fruits.update(["orange", "grapes"])
print(fruits)

fruits.remove("banana")   # Error if not found
print(fruits)
fruits.discard("banana")  # No error if not found
print(fruits)
fruits.pop()  # randomly pops any element
print(fruits)
fruits.remove("mango")           
print(fruits)
```
