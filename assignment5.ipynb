```python
import requests
def weather_data(city):
    url=f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid=811013a477bd1ea4de5aa66d89981035&&units=metric"
    try:
        r=requests.get(url)
        r.raise_for_status()
        d=r.json()
        print("city: ",d["name"])
        print("temperature: ",d["main"]["temp"],"°C")
        print("feel like:",d["main"]["feels_like"],"°C")  
        print("humidity:",d["main"]["humidity"],"%")
        print("Pressure:",d["main"]["pressure"])
        print("wearher:",d["weather"][0]["description"]) 
        print("wind speed:",d["wind"]["speed"])
        print(d['main']['temp'])
    except requests.exceptions.RequestException as t:
        print(t)
city=input("enter the city name:")
weather_data(city)   
      
```
