```python
#5) Iterate 3D Array Using For Loop and nditer
a=np.array([
    [[1,4,3,2],[2,54,2,6]],
    [[2,4,8,9],[3,45,6,3]]
            ])
print(a)
for i in a:
    for j in i:
        for k in j:
            print(k,end=" ")
for k in np.nditer(a):
    print(k)            
```

```python
#6) Study the following program
from scipy import stats
arr1 = np.array([3, 3,3])
arr2 = np.array([1, 1,1])
avg=(arr1+arr2)/2
print("Average of array",avg)
combine=np.concatenate((arr1.flatten(),arr2.flatten()))
print(combine)
print("mean=",np.mean(combine))
print("median=",np.median(combine))
print("mode= " ,stats.mode(combine,keepdims=True).mode[0])
```
