```python
import pandas as pd

data = {
    "Name": ["Lokesh", "Rahul"],
    "Marks": [90, 85]
}

df = pd.DataFrame(data)

new_row = pd.DataFrame({
    "Name": ["Aman"],
    "Marks": [88]
})

df = pd.concat([df.iloc[:1], new_row, df.iloc[1:]]).reset_index(drop=True)

print(df)
```
