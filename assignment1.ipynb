```python
mark1 = int(input("enter marks for subject 1: "))
mark2 = int(input("enter marks for subject 2: "))
mark3 = int(input("enter marks for subject 3: "))
mark4 = int(input("enter marks for subject 4: "))
mark5 = int(input("enter marks for subject 5: "))

total = int(mark1+mark2+mark3+mark4+mark5)
percentage = float(((mark1+mark2+mark3+mark4+mark5)/500)*100)
print("Percentage: ",percentage)
print("Total Marks: ",total)

if(percentage>=60.0):
    print("Grade:A")
elif(percentage<60.0 and percentage>=50.0):
    print("Grade=B")
elif(percentage<50.0 and percentage>=40.0):
    print("Grade=C")        
elif(percentage<40.0 and percentage>=33.0):
    print("Grade=D")    
else:
    print("Fail")    
```
