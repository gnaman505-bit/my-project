```python
#4)answer
t1=(4,5,1,7,8)
count=0
highest=t1[0]
lowest=t1[0]
for i in t1:
    count+=1
    if highest<i:
        highest=i
    if lowest>i:
        lowest=i
print("total:",count)
print("highest:",highest)
print("lowest:",lowest)            
    
    
    
```
